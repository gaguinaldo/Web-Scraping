```python
reg_4 = pd.DataFrame()
rule_list = []
for num in range(2, 94, 2):
    ipage = page_data.find_all('p')
    rule_num = ipage[num].find('a').text
    rule_url = ipage[num].find('a')['href']
    rule_name = ipage[num + 1].text
    rule_list.append(rule_num)
    
    
    reg_4 = reg_4.append(pd.DataFrame({'number': rule_num.split('\n')[0], 
                                       'part_url': rule_url, 
                                       'name': rule_name.split('\n')[0],
                                      'amend_date': rule_name[rule_name.find('(') + 1:rule_name.find(')')]}, index = [0]), ignore_index = True)
```

```python
reg_4
```

```python
reg_4['full_url'] = 'http://www.aqmd.gov' + reg_4['part_url']
```

```python
df_final = reg_4.drop('part_url', axis = 1)
```
